# Retrieve sentences and their corresponding tags

```python
class Sentencegetter(object):
    def __init__(self,data):
        self.n_sent=1
        self.data=data
        agg_func=lambda s :[(w,p,t) for w,p,t in zip(s["Word"].values.tolist(),s["POS"].values.tolist(),s["Tag"].values.tolist())] 
        self.grouped=self.data.groupby("Sentence #").apply(agg_func)
        self.sentences=[s for s in self.grouped]
```

```python
getter=Sentencegetter(data)
sentences=getter.sentences
```

```python
sentences[0]
```

# Define mappings between sentences and tags

```python
word2idx={w:i+1 for i,w in enumerate(words)}
tag2idx={t:i for i,t in enumerate(tags)}
```

```python
word2idx
```

```python
tag2idx
```

# Padding input sequences and creating train/test split

```python
plt.hist([len(s) for s in sentences],bins=50)
plt.show()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
max_len=50
X=[[word2idx[w[0]] for w in s] for s in sentences]
X=pad_sequences(maxlen=max_len,sequences=X,padding="post",value=num_words-1)
y=[[tag2idx[t[2]] for t in s ] for s in sentences ]
y=pad_sequences(maxlen=max_len,sequences=y,padding="post",value=tag2idx["O"])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=1)
```

# Building and Compiling a Bi-directional LSTM model

```python
from tensorflow.keras import Model , Input        # note : here we are not using sequential API but the functional API
from tensorflow.keras.layers import LSTM,Embedding,Dense
from tensorflow.keras.layers import TimeDistributed,SpatialDropout1D,Bidirectional
```

```python
input_word=Input(shape=(max_len,))
model=Embedding(input_dim=num_words,output_dim=max_len,input_length=max_len)(input_word)
model=SpatialDropout1D(0.1)(model)
model=Bidirectional(LSTM(units=100,return_sequences=True,recurrent_dropout=0.1))(model)
out=TimeDistributed(Dense(num_tags,activation="softmax"))(model)
model=Model(input_word,out)
model.summary()
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```
